```python
def loss_function(m,b,points):
  total_error=0
  for i in range(len(points)):
    x=points.iloc[i].Salary
    y=points.iloc[i].YearsExperience
    total_error+=(y-(m*i+b))**2
    return total_error/float(len(points))
```

```python
def gradient_descent(m_now,b_now,points,lr):
  m_gradient=0
  b_gradient=0

  n= len(points)

  for i in range(n):
    x=float(points.iloc[i].Salary)
    y=float(points.iloc[i].YearsExperience)

    m_gradient+= -(2/n)*x*(y-(m_now*x-b_now))
    b_gradient+= -(2/n)*(y-(m_now*x+b_now))

  m_now=m_now-m_gradient*lr
  b_now=b_now-b_gradient*lr

  return m_now,b_now
```

```python
# m=1/data['Salary'].max()
# b=data['YearsExperience'].mean()
m=0
b=0
L=0.0001
epochs=50

for i in range(epochs):
  m,b=gradient_descent(m,b,data,L)
  if(i%5==0):
    print(f"After {i+1} epochs:")
    print(m,b)
```
